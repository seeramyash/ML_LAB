```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/data.csv')
df.head()
```

```python
x_val = [0, 1, 2]
y_val = [3, 5, 9]
plt.plot(x_val, y_val)
```

```python
diagnosis_counts = df['diagnosis'].value_counts()
diagnosis_counts
```

```python
x_bar = diagnosis_counts.index
y_bar = diagnosis_counts
plt.bar(x_bar, y_bar)
```

```python
plt.figure(figsize=(12,8))
plt.bar(x_bar,y_bar)
```

```python
plt.figure(figsize=(12,8))
plt.bar(x_bar,y_bar)
plt.xticks(rotation=90, fontsize=10)
```

```python
plt.figure(figsize=(10,8))
plt.bar(x_bar,y_bar,width=0.2)
plt.title('Diagnosis Counts',fontsize=15)
plt.xlabel('Diagnosis',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.xticks(rotation=0, fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.bar(x_bar,y_bar,width=0.2,color='orange')
plt.title('Diagnosis Counts',fontsize=15)
plt.xlabel('Diagnosis',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.xticks(rotation = 0, fontsize=12)
plt.yticks(fontsize=12)
```

```python
plt.figure(figsize=(10,8))
plt.bar(x_bar,y_bar,width=0.2,color='orange')
plt.title('Diagnosis Counts',fontsize=15)
plt.xlabel('Diagnosis',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.xticks(rotation = 0, fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
sns.countplot(x='diagnosis', data=df, order=df['diagnosis'].value_counts().index, color='cornflowerblue')
plt.xticks(rotation=0)
plt.title('Diagnosis Counts', fontsize=15)
plt.xlabel('Diagnosis', fontsize=12)
plt.ylabel('Count', fontsize=12)
```

```python
diagnosis_counts = df['diagnosis'].value_counts()

plt.pie(diagnosis_counts,
        labels=diagnosis_counts.index,
        startangle=90,
        autopct='%.2f%%')

plt.title('Distribution of Diagnosis')
plt.show()
```
