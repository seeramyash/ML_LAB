```
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.metrics import r2_score # Import r2_score

# Reload the dataset to ensure correct 'diagnosis' values
df = pd.read_csv('data.csv')

# Drop the 'Unnamed: 32' column as it seems to contain mostly null values
if 'Unnamed: 32' in df.columns:
    df = df.drop('Unnamed: 32', axis=1)
    print("Dropped 'Unnamed: 32' column.")


# Ensure 'diagnosis' is in numerical format (M=1, B=0)
# Check unique values in 'diagnosis' before mapping
print("Unique values in 'diagnosis' column:", df['diagnosis'].unique())

# Remove rows where 'diagnosis' is not 'M' or 'B'
df_cleaned = df[df['diagnosis'].isin(['M', 'B'])].copy()

# Map 'diagnosis' to numerical format
df_cleaned['diagnosis'] = df_cleaned['diagnosis'].map({'M': 1, 'B': 0})


# Select 'radius_mean' as the feature and 'diagnosis' as the target
X_new = df_cleaned[['radius_mean']]  # Feature needs to be a 2D array
y_new = df_cleaned['diagnosis']

# Split the data into training and testing sets
X_train_new, X_test_new, y_train_new, y_test_new = train_test_split(X_new, y_new, test_size=0.2, random_state=42)

# Create a new linear regression model
new_model = LinearRegression()

# Train the new model
new_model.fit(X_train_new, y_train_new)

# Make predictions on the test set
y_pred_new = new_model.predict(X_test_new)

# Calculate the R-squared value for the new model
r2_new = r2_score(y_test_new, y_pred_new)

print("\nNew Linear Regression Model (using radius_mean to predict diagnosis):")
print("Coefficient:", new_model.coef_[0])
print("Intercept:", new_model.intercept_)
print(f"R-squared value for the new model: {r2_new}")
```
