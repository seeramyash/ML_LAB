```python
import numpy as np

# Sample arrays for demonstration
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
c = np.array([1, 2])
d = np.array([3, 4])
e = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
f = np.array([[[9, 10], [11, 12]], [[13, 14], [15, 16]]])

print("Dot product of a and b:\n", np.dot(a, b))
print("\nMulti-dot product of a, b, and a:\n", np.linalg.multi_dot([a, b, a]))
print("\nDot product of vectors c and d:\n", np.vdot(c, d))
print("\nInner product of a and b:\n", np.inner(a, b))
print("\nOuter product of c and d:\n", np.outer(c, d))
print("\nMatrix product of a and b:\n", np.matmul(a, b))
print("\nTensor dot product of e and f along specified axes:\n", np.tensordot(e, f, axes=([1, 0], [0, 1])))
print("\nEinstein summation (example - matrix multiplication a * b):\n", np.einsum('ij,jk->ik', a, b))

g = np.array([[1, 2], [3, 4]])
print("\nMatrix g raised to the power of 2:\n", np.linalg.matrix_power(g, 2))


h = np.array([[3, 1], [2, 2]])
eigenvalues, eigenvectors = np.linalg.eig(h)
print("\nEigenvalues of matrix h:", eigenvalues)
print("Right eigenvectors of matrix h:\n", eigenvectors)

symmetric_matrix = np.array([[1, 2], [2, 3]])
symmetric_eigenvalues, symmetric_eigenvectors = np.linalg.eigh(symmetric_matrix)
print("\nEigenvalues of symmetric matrix:\n", symmetric_eigenvalues)
print("Eigenvectors of symmetric matrix:\n", symmetric_eigenvectors)

general_matrix = np.array([[1, 1, 1], [2, 2, 2], [3, 3, 3]])
general_eigenvalues = np.linalg.eigvals(general_matrix)
print("\nEigenvalues of general matrix:\n", general_eigenvalues)
```
